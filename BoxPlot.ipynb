### BoxPlot
- sex = 0 --> male , 1 -----> Female ,
- smoker= 0 ----> no , 1 -----> Yes , 
- day = 3----> Thurs , 4----->Fri... , 
- time = 0 ----> Day , 1 -----> Night ,

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df= tips.copy()
df.head()
```

```python
df.describe().T
```

```python
df ["sex"].value_counts()
```

```python
df ["smoker"].value_counts()
```

```python
df ["day"].value_counts()
```

```python
df ["time"].value_counts()
```

----------------------

```python
sns.boxplot(x=df["total_bill"]);
```

```python
sns.boxplot (
    x="day",
    y="total_bill",
    data=df);
```

```python
sns.boxplot(
    x="time",
    y= "total_bill",
    data = df);
```

```python
sns.boxplot(
    x="time",
    y= "total_bill",
    hue="sex",
    data = df);
```
