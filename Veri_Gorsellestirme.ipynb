```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price, kde= False)
```

```python
sns.displot(df.price)
```

```python
sns.displot(df.price , bins = 3000);
```

```python
sns.kdeplot(df.price , fill=True);
```

```python
 (sns
  .FacetGrid(df,
             hue= "color",
             height = 5,
             xlim= (0,10000))
  .map(sns.kdeplot, "price" , fill = True)
  .add_legend()
 )
```

```python
(sns
  .FacetGrid(df,
             hue= "cut",
             height = 8,
             xlim= (0,10000))
  .map(sns.kdeplot, "price" , fill = False)
.add_legend()
)
```

```python
sns.displot(data=df, 
            x="price", 
            hue="cut", 
            kind="hist", 
            col="cut",
            col_wrap=5,
            height=7, );
```

```python
sns.catplot(
    x = "cut",
    y= "price",
    hue = "color",
    kind = "point",
    data= df);
```
